## _«Модель активации родопсина фотоном»_

### Коротко о родопсине

Один из способов преодолеть энергетический барьер между реагентами и продуктами и тем самым инициировать химические реакции — использовать энергию электромагнитного излучения. Реакции, происходящие под действием света, называют **_фотохимическими_**. Они происходят при взаимодействии вещества с видимым или УФ-излучением.

Одним из наиболее важных фотохимических процессов в биофизике выступает зрение. Первой стадией механизма зрения является поглощение видимого света белковым комплексом **_родопсином_**, который включает белок **_опсин_** и связанный с ним _хромофор_ **_ретиналь_**. _Хромофором_ называют молекулу или фрагмент молекулы, который поглощает видимый свет и тем самым обуславливает окраску вещества. 

Под действием света в родопсине происходит изомеризация ретиналя из _цис_- в _транс_-форму. В родопсине альдегидная группа     —CH=O ретиналя связана с одним из аминокислотных остатков белка, поэтому поворот одного фрагмента молекулы относительно другого при изомеризации приводит к изменению пространственного строения белка и появлению биологического сигнала - формированию нервного импульса. В результате поглощения кванта света молекулой родопсина инициируется цепь биохимических событий, следствием которой является **_закрытие катионных (Na +/Ca 2+) каналов_**, уменьшение темнового тока и гиперполяризация плазматической мембраны (увеличение положительного заряда на внешней поверхности клетки). Свет, повышая разность потенциалов на мембране рецепторной клетки (гиперполяризуя её), уменьшает выделение медиатора, фактически «выключая» рецепторы. 

Поглощение родопсином кванта света приводит к ряду его фотохимических превращений, первым этапом которых является **_фотоизомеризация_** _11-цис-ретиналя_ в полностью _транс_-форму. Изомеризация ретиналя, протекающая за очень короткое время в 200 фс, — единственный светозависимый процесс в зрительном цикле; остальные процессы сопряжены с конформационными перестройками в молекуле родопсина и являются следствием этого первичного акта поглощения света. Наибольшую важность для биохимических реакций, приводящих к возникновению фоторецепторного ответа, представляет один из интермедиатов фотолиза — _метародопсин II_ (максимум на спектре поглощения λmax=380 нм), который содержит полностью транс-ретиналь и характеризуется значительными конформационными перестройками по сравнению с «темновой» формой родопсина.

_Метародопсин II_ (R*, «световая» или «активированная» форма родопсина R) способен связывать особый G-белок, называемый трансдуцином, передающий возбуждение, сконцентрированное на молекуле трансмембранного рецептора, в жидкую фазу цитоплазмы. Трансдуцин (Т) относится к семейству гетеротримерных G-белков и состоит из трех субъединиц (Тα, Тβ и Тγ с массами 40, 37 и 8 кДа соответственно), в первой из которых (Тα) находится сайт связывания гуанидиновых нуклеотидов ГДФ и ГТФ. В темноте Тα-субъединица содержит связанную молекулу ГДФ и образует комплекс с субъединицами Тβ и Тγ (функционирующими как димер Тβγ). Комплекс (Тα–ГДФ)–Тβγ имеет высокое сродство к метародопсину II (R*), и после реакции с ним молекула связанного ГДФ обменивается на ГТФ. Комплекс R*–(Тα–ГТФ)–Тβγ быстро диссоциирует на R*, активный компонент Тα–ГТФ и димер Тβγ. Освобождённая молекула R* способна катализировать распад еще сотен или даже тысяч трансдуциновых комплексов, что является первым этапом усиления сигнала в процессе передачи зрительного сигнала.

Активный комплекс Тα–ГТФ, в свою очередь, активирует следующий белок зрительного каскада — гетеротетрамерную фосфодиэстеразу (ФДЭ) циклического ГМФ (цГМФ). Активация этого белка, локализованного на поверхности мембран дисков, приводит к расщеплению эффекторной молекулы — до трех тысяч молекул цГМФ на одну молекулу активного фермента — и инактивации цГМФ-зависимых катионных каналов, приводящей к гиперполяризации плазматической мембраны.

### Построение модели

За основу были взяты реакции, представляющие собой один из вариантов цикла активации/деактивации родопсина. Для простоты некоторые многостадийные реакции были заменены соответствующими одностадийными эффективными реакциями (например вне рассмотрения остались последовательные реакции фотоизомеризации Родопсин -> Метародопсин_1 -> Метародопсн_2 и т.д., а также реакции с участием трансдуцина). Для составления системы обратимся к закону действующих масс и закону Бугера-Ламберта-Бера. Где Xi - соответствующие концентрации веществ, ki - константы скоростей реакциий, I0 - интенсивность падающего излучения.


### Исследование модели

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# chemical reactions speed constants
k1 = 1000
k2 = 5
k3 = 0.1
k4 = 0.1
k5 = 1
k6 = 100

# time
t_max = 10

# initial conditions
X0 = 1, 0, 0, 0, 20, 0, 0

# additional conditions
def I(t):
    """
    Light intensity function.
    """
    return 1
```

```python
fig, ax = plt.subplots()

plt.plot(np.linspace(0, t_max), [I(element) for element in np.linspace(0, t_max)], color='r')

plt.ticklabel_format(axis="y", style="sci")
plt.xlabel('Time, arb. units')
plt.ylabel('Radiation intensity, arb. units')

ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='black', labelsize='medium', width=3)

plt.title('Dependence of light intensity on time')

plt.show()
```

```python
def derivatives(t, X):
    x1, x2, x3, x4, x5, x6, x7 = X
    x1_dot = -k1*I(t)*(1 - np.exp(-0.0005*x1)) + k6*x3*x7
    x2_dot = k1*I(t)*(1 - np.exp(-0.0005*x1)) - k2*x3*x4
    x3_dot = k2*x2 - k6*x3*x7
    x4_dot = k2*x2 - k3*x4*x5 + k4*x6
    x5_dot = -k3*x4*x5 + k4*x6 + k5*x6
    x6_dot = k3*x4*x5 - k4*x6 - k5*x6
    x7_dot = k5*x6 - k6*x3*x7
    return x1_dot, x2_dot, x3_dot, x4_dot, x5_dot, x6_dot, x7_dot

solution = solve_ivp(derivatives, 
                     (0, t_max), X0, 
                     method='Radau', # Implicit Runge-Kutta method of Radau IIA family of order 5
                     max_step=0.1)

print(solution)
```

```python
fig, ax = plt.subplots()

plt.plot(solution.t, solution.y[0], 
         marker='o', label='Rhodopsin', 
         color="r", linestyle='--')

plt.plot(solution.t, solution.y[1], 
         marker='o', label='Activated rhodopsin', 
         color="b", linestyle='--')

plt.ticklabel_format(axis="y", style="sci")
plt.xlabel('Time, arb. units')
plt.ylabel('Concentration, arb. units')

ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='black', labelsize='medium', width=3)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Kinetics of rhodopsin activation/deactivation cycle')

plt.show()
```
